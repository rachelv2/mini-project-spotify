```python
# Format bridge table track_genre
import pandas as pd
import numpy as np
import ast 
# import dataset
df = pd.read_csv(r"C:\Users\Utilisateur\Documents\spotify\charts.csv", encoding="ISO-8859-1", sep=";")
df
```

```python
# date filter on the last 5 years and convert into year
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
df = df.loc[df['date'] > '2020-12-31'].copy()
df['date'] = df['date'].dt.year
print(df.head())
```

```python
# return string as a list
import ast

def safe_eval_genre(s):
    try:
        return ast.literal_eval(s)
    except:
        return []

df['artist_genres_list'] = df['artist_genres'].apply(safe_eval_genre)
```

```python
df['artist_genres_list']
```

```python
genre_map = {
    "pop": ["pop", "pop rap", "pop dance", "dance pop", "electropop", "uk pop", "viral pop", "canadian pop"],
    "hip hop/rap": ["hip hop", "rap", "trap", "pop rap", "latin hip hop", "southern hip hop", "gangster rap"],
    "electronic/dance": ["edm", "electro house", "house", "progressive house", "tropical house", "trance", "complextro"],
    "latin": ["reggaeton", "bachata", "cumbia", "latin pop", "urbano latino"],
    "r&b/soul": ["r&b", "neo soul", "contemporary r&b", "chill r&b", "pop r&b"],
    "rock": ["rock", "pop rock", "modern rock", "alternative rock", "indie rock"],
    "country/folk": ["country", "country pop", "folk pop"],
    "global/regional": ["k-pop", "j-pop", "mandopop", "italian pop", "french pop"],
    "alternative/indie": ["indie pop", "indie poptimism", "bedroom pop", "alt z", "art pop"],
    "reggae/dancehall": ["dancehall", "moombahton", "tropical"]
}
```

```python
def assign_main_genre(genres):
    if not genres:
        return "Other"
    
    genres = [g.lower() for g in genres]
    
    for main_genre, subgenres in genre_map.items():
        for sub in subgenres:
            if any(sub == g or sub in g for g in genres):
                return main_genre.capitalize()
    return "Other"
```

```python
df['main_genre'] = df['artist_genres_list'].apply(assign_main_genre)
```

```python
genres_df = pd.DataFrame({"genre_id": [i for i in range(1, len( list( genre_map.keys() ) ) +1)],"main_genre": list(genre_map.keys())})
genres_df
```

```python
# rename main_genre and export as a csv
genres_df = genres_df.rename(columns={'main_genre': 'genre_name'})
genres_df.to_csv("genre.csv", index=False, encoding="utf-8", sep=";")
```

```python
# same case
df['main_genre'] = df['main_genre'].str.lower()
genres_df['genre_name'] = genres_df['genre_name'].str.lower()

# merge
df = df.merge(genres_df, left_on='main_genre', right_on='genre_name', how='left')
```

```python
# kepp the columns we want
track_genre = df[['track_id', 'genre_id']].copy()
track_genre = track_genre.rename(columns={'track_id': 'trackID'})

# drop duplicates
track_genre = track_genre.drop_duplicates().reset_index(drop=True)

# auto id
track_genre.insert(0, 'id', range(1, len(track_genre) + 1))

# csv 
track_genre.to_csv("track_genre.csv", index=False, encoding="utf-8", sep=";")
```

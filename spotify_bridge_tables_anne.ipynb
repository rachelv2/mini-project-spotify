```python
# Format bridge table track_genre
import pandas as pd
import numpy as np
import ast 
# import dataset
df = pd.read_csv(r"C:\Users\Utilisateur\Desktop\charts.csv", encoding="ISO-8859-1", sep=";")
# convert string into a list
df["genre"] = df["artist_genres"].apply(ast.literal_eval)
# explode into rows each genre
df_exploded = df.explode("genre")
# keep specific columns and rename them
track_genre = df_exploded[["track_id", "genre"]].rename(columns={"track_id": "trackID", "genre": "genre_name"}).copy().reset_index(drop=True)
# add an autoincrement id 
track_genre.insert(0, "id", range(1, len(track_genre) + 1))
# export into a csv file
track_genre.to_csv("C:/Users/Utilisateur/Desktop/filled_full_tables/track_genre.csv", index=False)
```
